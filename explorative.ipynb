```python
import pandas as pd


df = pd.read_csv("taxonomy.csv")
df["Type"] = df["Type"].apply(lambda x: x.strip())
```

```python
import matplotlib.pyplot as plt


grouped = df.groupby(["Language", "Category"]).size().unstack(fill_value=0)

# Plot
grouped.plot(kind="bar", stacked=True, figsize=(10, 7), colormap="tab20b")
plt.xticks(rotation=45)
plt.ylabel("Number of Task Categories")
# plt.title('Number of Task Categories per Language')
plt.legend(
    title="Task Category",
    loc="lower center",
    bbox_to_anchor=(0.5, 1.05),
    ncol=3,
    fancybox=True,
    shadow=True,
)
plt.tight_layout()
plt.show()
```
